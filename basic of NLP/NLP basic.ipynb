## 육아휴직 관련 법안 분석
- 형태소 분석

```python
import nltk
from konlpy.corpus import kobill

doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko
```

```python
from konlpy.tag import Okt
```

```python
t = Okt()
tokens_ko  = t.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name = '육아휴직법안')
ko
```

```python
# 단어의 전체 개수
len(ko.tokens)
```

```python
# 중복을 제외하고 한번 이상씩 나온 단어
len(set(ko.tokens))
```

```python
# 그래프로 표현
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.figure(figsize = (12,6))
#가장 많이 나온 단어 50개만 출력
ko.plot(50);
```

```python
# stopwords 만들기
# 한글의 stopwords는 상황에 따라 복잡해서, 손으로 일단 잡음
stop_words = [
    ".", ",", "(", ")", "'", '"', "·", "-", "_", "…", "!", "?", ":", ";", 
    "의", "자", "에", "안", "번", "호", "을", "이", "다", "만", "로", "가", "를"
]
```

```python
ko = [each_word for each_word in ko if each_word not in stop_words]
```

```python
ko.count('초등학교')
```

```python
ko  = nltk.Text(ko, name ='대한민국 국회의안 제 1809890호')
plt.figure(figsize=(12,6))
ko.plot(50);
```

- dispersion_plot 메서드는 일반적으로 NLTK 라이브러리의 Text 객체에서 사용하는 메서드
- 이 메서드를 사용하려면 먼저 텍스트 데이터를 nltk.Text 객체로 변환해야 한다.

```python
plt.figure(figsize=(15,6))

#전체 문장에서 아래의 단어들이 어디에 위치하고 있는지 보여줌.
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
# 해당 단어 좌우에 있는 글자들까지 보여줌
ko.concordance(['초등학교'])
```

**영어에서 연어(連語, collocation)란 무엇인가?**

연어(連語, collocation)란 "함께 위치하는 단어들"(co + location)이란 뜻으로, 어휘의 조합 또는 짝을 이루는 말들을 일컫습니다. 예를 들면 'friend'와 주로 같이 쓰이는 형용사는 'best', 'good', 그리고 'loyal'입니다. 예문은 다음과 같습니다.

"She talks to her best friend every day."
(best)

출처: http://ko.talkenglish.com/how-to-use/collocations.aspx

```python
# 아래 collocation이 실행되지 않을때 아래 코드 실행
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
print(stop_words)
```
